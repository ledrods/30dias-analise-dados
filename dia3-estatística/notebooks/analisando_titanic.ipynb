```python
#importando bibliotecas
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# primeiro passo é entender os dados
# Carregar o dataset Titanic
df = pd.read_csv("L:/usuários/leandro/OneDrive/Área de Trabalho/cronograma30/cronograma-analise-dados/dia3-estatística/data/titanic.csv")

# Mostrar as primeiras linhas
print(df.head())

# Informações gerais do dataset
print(df.info())

# Ver estatísticas básicas
print(df.describe())
```

```python

plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)  # KDE mostra a suavização da distribuição
plt.title('Distribuição das Idades dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
"""
Esse gráfico nos mostra se há um equilibrío entre a quantidade de
jovens e idosos à bordo do Titanic.
"""
```

Pelo gráfico de distribuição das idades dos passageiros podemos ver que a maior parte dos passageiros 
eram pessoas com menos de 35 anos, pois a partir desse número a frequência de passageiros cai muito

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='viridis')
plt.title('Taxa de Sobrevivência por Classe')
plt.xlabel('Classe')
plt.ylabel('Quantidade de Passageiros')
plt.legend(['Não Sobreviveu', 'Sobreviveu'])
plt.show()
```

Com esse gráfico da taxa de sobrevivência por classe podemos perceber que a classe dos passageiros teve grande influência na sobrevivência
Passageiros da primeira classe tiveram mais chance de sobreviver, enquanto os da 3º classe foram os mais afetados.
