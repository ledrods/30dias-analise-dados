```python
# importando bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# pequeno conjunto de dados para praticar média, mediana, moda, variancia, desvio padrão e amplitude

data = {'Aluno': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
        'Altura (cm)': [165, 170, 160, 175, 168, 172, 163, 178, 165, 170]}

import pandas as pd
df = pd.DataFrame(data)

media_estudantes = df['Altura (cm)'].mean()

mediana_Estudantes = df['Altura (cm)'].median()

moda_Estudantes = df['Altura (cm)'].mode()

variancia_Estudantes = df['Altura (cm)'].var()

desvio_padrao_Estudantes = df['Altura (cm)'].std()

amplitude_Estudantes = df['Altura (cm)'].max() - df['Altura (cm)'].min()

print("A média de altura dos estudantes é", media_estudantes)
print("a mediana de altura dos estudantes é", mediana_Estudantes)
print("a moda de altura entre os estudantes é:", moda_Estudantes)
print("A variancia de altura dos alunos é:", variancia_Estudantes)
print("O desvio padrão de altura dos alunos é:", desvio_padrao_Estudantes)
print("a amplitude entre os estudantes é:", amplitude_Estudantes)
```

```python
#Histograma das Alturas Originais
alturas = np.array([165, 170, 160, 175, 168, 172, 163, 178, 165, 170])

plt.hist(alturas, bins=5, density=True, alpha=0.6)
```

```python
#Distribuição Normal Baseada nos Dados
media = np.mean(alturas)
std = np.std(alturas)

normal = np.random.normal(media, std, size=1000)

plt.hist(normal, bins=5, density=True, alpha=0.6)
```

```python
#Distribuição Binomial
alturas = np.array([165, 170, 160, 175, 168, 172, 163, 178, 165, 170])

n = len(alturas)

p = np.sum(alturas > 170) / n

binomial = np.random.binomial(n, p, size=1000)


plt.hist(binomial, bins=5, density=True, alpha=0.6)
```
